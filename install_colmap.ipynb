```python
conda create -n MyEnvForColmap python=3.10.10
conda config --add channels conda-forge
conda config --set channel_priority strict
conda install mamba
mamba install \
    cmake \
    ninja \
    boost \
    ccache \
    eigen \
    openimageio \
    curl \
    metis \
    glog \
    gtest \
    ceres-solver \
    suitesparse \
    qt \
    glew \
    sqlite \
    cgal-cpp \
    mesa-libgl-devel-cos7-x86_64 \
    cuda-compiler==12.6.2 \
    cuda-cudart-dev \
    cuda-nvrtc-dev \
    libcurand-dev

git clone https://github.com/colmap/colmap.git
cd colmap
mkdir build
cd build

cmake .. -GNinja -DGUI=OFF \
  -DOpenMP_C_FLAGS="-fopenmp" \
  -DOpenMP_CXX_FLAGS="-fopenmp" \
  -DOpenMP_C_LIB_NAMES="gomp" \
  -DOpenMP_CXX_LIB_NAMES="gomp" \
  -DOpenMP_gomp_LIBRARY=$CONDA_PREFIX/lib/libgomp.so \
  -DCMAKE_EXE_LINKER_FLAGS="-L$CONDA_PREFIX/lib"

ninja

```
