```python
# use vscode to change windows line end to lf
1. 在 VS Code 中转换换行符 (推荐)
VS Code 右下角的状态栏可以直接修改文件的换行格式：

在 VS Code 中打开 pytorch.sh。

观察右下角状态栏，你会看到 CRLF（这是 Windows 格式）。

点击 CRLF，在顶部弹出的菜单中选择 LF。

保存文件 (Ctrl + S)。

重新上传或同步到 AI Studio 即可。
```
