```python
import trimesh

# Path to your GLB file
file_path = '/home/aistudio/GLTF/MyBicycles.glb'

# 1. Load the file
# Trimesh automatically detects this as a Scene or a Mesh
scene_or_mesh = trimesh.load(file_path)

# 2. Check if it's a Scene (multiple parts) or a single Mesh
if isinstance(scene_or_mesh, trimesh.Scene):
    print(f"Loaded a Scene with {len(scene_or_mesh.geometry)} objects.")
    # To get a single merged mesh of the whole bicycle:
    mesh = scene_or_mesh.dump(concatenate=True)
else:
    mesh = scene_or_mesh
    print("Loaded a single Mesh object.")

# 3. Print basic stats
print(f"--- MyBicycles Stats ---")
print(f"Vertices: {len(mesh.vertices)}")
print(f"Faces (Polygons): {len(mesh.faces)}")
print(f"Bounding Box: {mesh.bounds[1] - mesh.bounds[0]}")
print(f"Is it watertight? {mesh.is_watertight}")
```
